```python
import pandas as pd
df=pd.read_csv('titanic.csv')
df.head()
```

```python
print(df.isnull().sum())
```

```python
def fill_missing_with_mean(df, columns_to_check):
    for column in columns_to_check:
        if df[column].isnull().sum() > 0:
            df[column].fillna(df[column].mean(), inplace=True)
    return df
columns_with_zero = ['PassengerId', 'Survived']
df = fill_missing_with_mean(df, columns_with_zero)
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')
plt.figure(figsize=(10, 6))
sns.scatterplot(data=titanic, x='age', y='fare',color='blue', alpha=0.7)
plt.title('Scatterplot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
# Histogram of Age and Fare
df['Age'].hist(bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

df['Fare'].hist(bins=30, color='salmon', edgecolor='black', alpha=0.7)
plt.title('Distribution of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```
